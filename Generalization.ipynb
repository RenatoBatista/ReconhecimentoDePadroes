```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from google.colab import files
data = files.upload()
#data = pd.read_csv('../inputs/classification/smoke_detection_iot.csv')
#site: https://www.kaggle.com/datasets/karthickveerakumar/salary-data-simple-linear-regression
```

```python
import io
data = pd.read_csv(io.BytesIO(data['smoke_detection_iot.csv']))
```

```python
data.describe()
```

```python
X_, y_ = data.drop('Fire Alarm',axis=1), data['Fire Alarm']
X_.shape, y_.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, make_scorer, accuracy_score,f1_score
from sklearn.model_selection import cross_validate
modelo = KNeighborsRegressor()
scores_mse = cross_validate(modelo, X_, y_, scoring=make_scorer(mean_squared_error, squared=False),cv=10)
scores_acc = cross_validate(modelo, X_, y_, scoring=make_scorer(accuracy_score),cv=10)
print(scores_mse['test_score'], "Média: ", scores_mse['test_score'].mean())
print(scores_acc['test_score'], "Média: ", scores_mse['test_score'].mean())
```

```python
modelo = LogisticRegression()
scores_mse = cross_validate(modelo, X_, y_, scoring=make_scorer(mean_squared_error, squared=False),cv=10)
scores_acc = cross_validate(modelo, X_, y_, scoring=make_scorer(accuracy_score),cv=10)
print(scores_mse['test_score'], "Média: ", scores_mse['test_score'].mean())
print(scores_acc['test_score'], "Média: ", scores_mse['test_score'].mean())
```
